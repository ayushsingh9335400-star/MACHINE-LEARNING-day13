# MISSING indicator / random sample imputation/ aautomatice selection

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 

from sklearn.model_selection import train_test_split

import seaborn as sns 
```

```python
df = pd.read_csv('titanic_toy.csv',usecols=['Age','Fare','Survived'])
```

```python
df.head()
```

```python
df.isnull().mean()*100
```

```python
X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)

X_train
```

```python
X_train['Age_imputrd'] = X_train['Age']
X_test['Age_imputed'] = X_test['Age']
```

```python
X_test.head()
```

```python
import numpy as np

# Impute missing values in X_train
X_train.loc[X_train['Age_imputed'].isnull(), 'Age_imputed'] = (
    X_train['Age'].dropna()
    .sample(X_train['Age_imputed'].isnull().sum(), replace=True, random_state=42)
    .values
)

# Impute missing values in X_test
X_test.loc[X_test['Age_imputed'].isnull(), 'Age_imputed'] = (
    X_train['Age'].dropna()   # always use training distribution
    .sample(X_test['Age_imputed'].isnull().sum(), replace=True, random_state=42)
    .values
)
```
